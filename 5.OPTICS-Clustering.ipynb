```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,3:5].values
```

```python
x
```

```python
from sklearn.cluster import OPTICS
optics = OPTICS(min_samples=7, min_cluster_size=5)  # Tune 'eps' and 'min_samples' as needed
optics.fit(x)
y_optics = optics.fit_predict(x)
```

```python
y_optics
```

```python
dir(optics)
```

```python
# Get cluster labels and reachability distances
labels = optics.labels_
reachability = optics.reachability_
ordering = optics.ordering_
```

```python
dataset['Cluster'] = y_optics
```

```python
# Plot the Reachability Plot
plt.figure(figsize=(10, 5))
plt.plot(range(len(reachability)), reachability[ordering], marker='o', linestyle='-', color='b')
plt.title("Reachability Plot for OPTICS Clustering")
plt.xlabel("Sample Index (Ordered)")
plt.ylabel("Reachability Distance")
plt.show()
```

```python
# Plot the clustered data
plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("OPTICS Clustering on Mall Customers Data")
plt.colorbar(label="Cluster")
plt.show()
```
